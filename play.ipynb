```python
import numpy as np
import robosuite as suite
import matplotlib.pyplot as plt

# create environment instance
env = suite.make(
    env_name="TwoArmTransport", # try with other tasks like "Stack" and "Door"
    robots="PlaifDualRB3",  # try with other robots like "Sawyer" and "Jaco"
    has_renderer=True,
    has_offscreen_renderer=True,
    use_camera_obs=True,
    camera_names=["robot0_stand_belly_cam", "robot0_stand_head_cam"],
    camera_heights=[1080, 720], 
    camera_widths=[1920, 1280],
)
```

```python
# reset the environment
obs, done = env.reset(), False
print(env.action_spec)

j = 0
while not done:
    action = np.random.randn(env.robots[0].dof) # sample random action
    obs, reward, done, info = env.step(action)  # take action in the environment
    env.render()  # render on display
    j += 1
    if j > 10:
        break
```

```python
# show obs['robot0_stand_belly_cam_image'] and obs['robot0_stand_head_cam_image'] side by side
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(obs['robot0_stand_belly_cam_image'])
plt.title('robot0_stand_belly_cam_image')
plt.subplot(1, 2, 2)
plt.imshow(obs['robot0_stand_head_cam_image'])
plt.title('robot0_stand_head_cam_image')
plt.show()
```
